```python
n_samples = 5000  # Сколько строк генерить
min_skills = 3    # Минимум навыков в одной записи
max_skills = 12   # Максимум навыков

# Генерация данных
generated_data = []

for _ in range(n_samples):
    # Выбираем случайную профессию
    profession = random.choice(list(professions_skills_expanded_ru.keys()))
    true_skills = professions_skills_expanded_ru[profession]

    # Выбираем сколько реальных навыков взять
    n_true_skills = random.randint(int(min_skills * 0.7), min(len(true_skills), max_skills - 2))

    # Выбираем реальные навыки (рандомно)
    selected_true_skills = random.sample(true_skills, n_true_skills)

    # Добавляем "шумовые" навыки
    n_noise_skills = random.randint(1, 3)
    selected_noise_skills = random.sample(common_skills, n_noise_skills)

    # Собираем итоговый список навыков
    final_skills = selected_true_skills + selected_noise_skills
    random.shuffle(final_skills)

    # Сохраняем запись
    generated_data.append({
        "skills": ", ".join(final_skills),
        "profession": profession
    })
```

```python
df_skills_noisy = pd.DataFrame(generated_data)

# Сохраняем в CSV
df_skills_noisy.to_csv('skills.csv', index=False, encoding='utf-8')
```
